```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import imutils
from imutils.perspective import four_point_transform
blue = (255,0,0)
green = (0,255,0)
```

```python
# Loading and Displaying Image
path = "C:/Users/HP/Desktop/MS/Semester 2/Computer Vison/"
img = cv2.imread("C:/Users/HP/Desktop/MS/Semester 2/Computer Vison/5.jpg")
orig_image = img.copy()
img = cv2.resize(img, (600,800))
```

```python
# Blurring and Edge Detection
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert the image to gray scale
blur = cv2.GaussianBlur(gray, (5, 5), 0) # Add Gaussian blur
cv2.imwrite(path + "result1blur.jpg", blur)
edged = cv2.Canny(blur, 75, 200) # Apply the Canny algorithm to find the edges
cv2.imwrite(path + "result2canny.jpg", edged)
```

```python
# Getting Contours
(contours, _) = cv2.findContours(edged.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
contours = sorted(contours, key=cv2.contourArea, reverse=True)
cv2.drawContours(img, contours, -1, blue, 3)
cv2.imwrite(path + "result3drawcountour.jpg", img)  
```

```python
# Selecting Highest Contour
# go through each contour
for c in contours:
    # Approximating each contour
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.02 * peri, True)
    # if found a countour with 4 points break the for loop
    # Assuming document is found
    if len(approx) == 4:
        count = approx
        break
# Drawing the contours on the original image
cv2.drawContours(img, [count], -1, green, 3)
cv2.imwrite(path + "result4selectedcountour.jpg", img)  
# apply warp perspective to get the top-down view
warped = four_point_transform(orig_image, count.reshape(4, 2))
# convert the warped image to grayscale
warped = cv2.cvtColor(warped, cv2.COLOR_BGR2GRAY)
warped = imutils.resize(warped,height=800)
cv2.imwrite(path + "result5Final.jpg", warped)
```

```python
imgs = cv2.imread("result4selectedcountour.jpg")
rows,cols,ch = imgs.shape
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(imgs,M,(500,500))
plt.subplot(121),plt.imshow(imgs),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
